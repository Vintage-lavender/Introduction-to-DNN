4. Pytorch로 hidden layer 10개짜리 neural network과 hidden layer 2개짜리 neural network을 이용하여 XOR를 학습시키는 코드를 각각 작성하여 학습시키시오. 학습 결과를 제시하고 그 이유를 설명하시오. 각 layer의 node 개수는 4개이다. Binary class classification 문제를 기준으로 neural network를 구성하시오. (코드, 결과, 설명을 함께 제시하시오.)

```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim

#training data
x_train = torch.FloatTensor([[1,1],
                              [1,0],
                              [0,1],
                              [0,0]])  #XOR
                              
y_train = torch.FloatTensor([[0],
                              [1],
                              [1],
                              [0]])
```

```python
#hidden layer 10개짜리 neural network
class ten_MRModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 4, bias=True )
        #10개의 은닉층: 각 층마다 노드 개수 4개씩
        self.linear1 = nn.Linear(4, 4)
        self.linear2 = nn.Linear(4, 4)
        self.linear3 = nn.Linear(4, 4)
        self.linear4 = nn.Linear(4, 4)
        self.linear5 = nn.Linear(4, 4)
        self.linear6 = nn.Linear(4, 4)
        self.linear7 = nn.Linear(4, 4)
        self.linear8 = nn.Linear(4, 4)
        self.linear9 = nn.Linear(4, 4)
        self.linear10 = nn.Linear(4, 4)
        self.outlinear = nn.Linear(4, 1)
        self.ReLU = nn.ReLU() #hidden layer activation function : ReLU
        self.sigmoid = nn.Sigmoid() #output layer activation function : sigmoid
        
    def forward(self, x):
        #10개의 은닉층에서 사용하는 활성화함수
        x = self.ReLU(self.linear(x))
        x = self.ReLU(self.linear1(x))
        x = self.ReLU(self.linear2(x))
        x = self.ReLU(self.linear3(x))
        x = self.ReLU(self.linear4(x))
        x = self.ReLU(self.linear5(x))
        x = self.ReLU(self.linear6(x))
        x = self.ReLU(self.linear7(x))
        x = self.ReLU(self.linear8(x))
        x = self.ReLU(self.linear9(x))
        x = self.ReLU(self.linear10(x))
        x = self.sigmoid(self.outlinear(x))
        return x 

#hidden layer 10개짜리 neural network model 객체 생성
ten_model = ten_MRModel()

#Loss funtion = Cross Entropy
criterion = torch.nn.BCELoss()
optimizer = optim.SGD(ten_model.parameters(), lr=0.1)

#학습: 데이터셋으로 이루어지는 학습횟수 = 50000번
nb_epochs = 50000

for epoch in range(nb_epochs + 1):
    # H(x) 계산
    prediction = ten_model(x_train) #학습에 사용된 입력값의 출력값을 모델로 예측
    # cost 계산 
    cost = criterion(prediction, y_train) #target값과 예측한 값의 오차를 CE로 계산
    # cost로 H(x) 개선   
    optimizer.zero_grad() #gradient 초기화
    cost.backward() #역전파
    optimizer.step()  #connection weight update
    
    if epoch % 1000 == 0 :  
        print('Epoch {:4d}/{} Cost: {:.6f}'.format( epoch, nb_epochs, cost.item() ))
        #print(ten_model.state_dict() )
```

```python
#hidden layer 2개짜리 neural network
class two_MRModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 4, bias=True )
        #2개의 은닉층: 각 층마다 노드 개수 4개씩
        self.linear1 = nn.Linear(4, 4)
        self.linear2 = nn.Linear(4, 4)
        self.outlinear = nn.Linear(4, 1)
        self.ReLU = nn.ReLU() #hiddenlayer activation function : ReLU
        self.sigmoid = nn.Sigmoid() #output layer activation function : sigmoid
        
    def forward(self, x):
        x = self.ReLU(self.linear(x))
        #2개의 은닉층에서 사용하는 활성화함수
        x = self.ReLU(self.linear1(x))
        x = self.ReLU(self.linear2(x))
        x = self.sigmoid(self.outlinear(x))
        return x 
  
#hidden layer 2개짜리 neural network model 객체 생성
two_model = two_MRModel()

#Loss funtion = Cross Entropy
criterion = torch.nn.BCELoss()
optimizer = optim.SGD(two_model.parameters(), lr=0.1)

#학습: 데이터셋으로 이루어지는 학습횟수 = 50000번
nb_epochs = 50000

for epoch in range(nb_epochs + 1):
    # H(x) 계산
    prediction = two_model(x_train) #학습에 사용된 입력값의 출력값을 모델로 예측
    # cost 계산
    cost =criterion(prediction, y_train) #target값과 예측한 값의 오차를 CE로 계산
    # cost로 H(x) 개선
    optimizer.zero_grad() #gradient 초기화
    cost.backward() #역전파
    optimizer.step()  #connection weight update
    
    if epoch % 1000 == 0 :  
        print('Epoch {:4d}/{} Cost: {:.6f}'.format( epoch, nb_epochs, cost.item() ))
        #print(two_model.state_dict() )
```
